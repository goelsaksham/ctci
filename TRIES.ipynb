```python
class Node:
    def __init__(self, val = ''):
        self.val = val
        self.children = {}
        
    def add_children(self, char, node):
        self.children[char] = node
        
    def get_children(self, char):
        if char in self.children:
            return self.children[char]
        else:
            return None
        

        
def add_to_trie(root, word):
    current_node = root
    for char in word:
        children_node = current_node.get_children(char)
        if not children_node:
            current_node.add_children(char, Node())
        current_node = current_node.get_children(char)
    current_node.val = word
    
def find_in_trie(root, word):
    current_node = root
    for char in word:
        child_node = current_node.get_children(char)
        if not child_node:
            return False
        current_node = current_node.get_children(char)
    return not (current_node.val == '')

def delete_trie_node(root, word):
    def helper(node, remaining_word):
        if remaining_word:
            current_char = remaining_word[0]
            remaining_word = remaining_word[1:]
            if current_char in node.children:
                if helper(node.children[current_char], remaining_word):
                    node.children.pop(current_char)
                    return len(node.children) == 0
                else:
                    return False
            else:
                return False
        else:
            root.val = ''
            return (node != None) and (len(node.children) == 0)
        
    current_char = word[0]
    remaining_word = word[1:]
    if current_char in root.children:
        if helper(root.children[current_char], remaining_word):
            root.children.pop(current_char)
            
def print_trie(root):
    if root.val != '':
        print(root.val)
    for children_chr in root.children:
        print_trie(root.children[children_chr])
    
my_trie = Node()
add_to_trie(my_trie, 'aardvark')
add_to_trie(my_trie, 'bad')
add_to_trie(my_trie, 'badder')
add_to_trie(my_trie, 'sad')
#print(find_in_trie(my_trie, 'aardvark'))
#print(find_in_trie(my_trie, 'aardvar'))
#print(find_in_trie(my_trie, 'bad'))
#print(find_in_trie(my_trie, 'badder'))
#print(find_in_trie(my_trie, 'baddest'))
delete_trie_node(my_trie, 'sad')
delete_trie_node(my_trie, 'bad')

print_trie(my_trie)
```
